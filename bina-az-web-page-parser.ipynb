```python
import requests  
from bs4 import BeautifulSoup
import pandas as pd
```

```python
link = 'https://bina.az'
```

```python
source = requests.get(link)
source.text
```

```python
soup = BeautifulSoup(source.text, 'html.parser')
soup
```

```python
soup.find_all('div', {'class': 'location'})
```

```python
location = []

for i in soup.find_all('div', {'class': 'location'}):
    location.append(i.text)
    
location
```

```python
soup.find_all('div', {'class': 'items-i featured vipped'})
```

```python
url = []

for i in soup.find_all('a', {'class': 'item_link'}):
    url.append('https://bina.az' + i.get('href'))
    
unique_url = list(dict.fromkeys(url))
len(unique_url)

unique_url
```

```python
soup.find_all('div', {'class': 'card_footer'})
```

```python
date = []

for i in soup.find_all('div', {'class': 'card_footer'}):
    date.append(i.find('div').text)

date
```

```python
soup.find_all('div', {'class': 'abs_block'})
```

```python
price = []

for i in soup.find_all('div', {'class': 'abs_block'}):
    price.append(i.find('div').text)

price 
```

```python
pd.DataFrame({'Location': location, 'Link': unique_url, 'Date': date, 'Price': price})
```
